```python
import matplotlib.pyplot as plt # Import matplotlib for plotting
from sklearn.svm import SVR
import numpy as np


# Generate sample data for demonstration purposes
rng = np.random.RandomState(0)
x = 5 * rng.rand(100, 1) # Input features
y = np.sin(x).ravel() # Target variable
y[::5] += 3 * (0.5 - rng.rand(20)) # Add noise to some targets
line = np.linspace(0, 5, 100).reshape(-1, 1)  # Values for plotting predictions
k = 'darkorange' # Color for plotting data points

for gamma in[1, 10]:
  svr = SVR(gamma=gamma).fit(x,y)
  plt.plot(line, svr.predict(line), label = 'SVR gamma ={}'. format(gamma))

plt.plot(x[:, 0], y, 'o', c=k)
plt.xlabel("Regression output")
plt. ylabel("Input feature")
plt.legend(loc="best")
```
